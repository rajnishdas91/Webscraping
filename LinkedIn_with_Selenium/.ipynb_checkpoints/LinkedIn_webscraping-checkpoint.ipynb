```python
# Imort the necessary libraries
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By

# URL to scrape
url = "https://www.linkedin.com/jobs/search/?currentJobId=3910363798&f_C=17955831&geoId=92000000&origin=COMPANY_PAGE_JOBS_CLUSTER_EXPANSION&originToLandingJobPostings=3910363798%2C3909817359%2C3910368050%2C3910367249%2C3909816443%2C3910368048%2C3910368049%2C3909815943%2C3867098260"

# Configure Chrome options
chrome_options = Options()
chrome_options.add_argument("--headless")  # Run Chrome in headless mode (without opening GUI)
chrome_options.add_argument("--start-maximized")  # Maximize the window when opening

# Initialize Chrome WebDriver
driver = webdriver.Chrome(options=chrome_options)

# Open the URL
driver.get(url)

# Find all job titles and their links
job_elements = driver.find_elements(By.CLASS_NAME, "base-card__full-link")

# Extract and print job titles and their links
for job_element in job_elements:
    job_title = job_element.find_element(By.CLASS_NAME, "sr-only").text
    job_link = job_element.get_attribute("href")
    print(f"Job Title: {job_title}")
    print(f"Job Link: {job_link}")
    print()

# Close the browser
driver.quit()
```
